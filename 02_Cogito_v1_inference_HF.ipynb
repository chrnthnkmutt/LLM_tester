```python
vocab = data['model']['vocab']
type(vocab), len(vocab)
```

```python
import re
```

```python
pat = re.compile('[ก-๙]')
```

```python
thais = [v for v in vocab.keys() if pat.search(v)]
len(thais)
```

```python
','.join(sorted(thais))
```

```python
# ตัวอักษรไทยที่ควรมี
thai_chars = []
pat2 = re.compile('[ก-ฮะ-ูเ-์๐-๙]')
for i in range(ord('ก'), ord('๙')+3):
  if pat2.search(chr(i)):
    thai_chars.append(chr(i))
len(thai_chars)
```

```python
# ตัวที่ขาดไป  ต้องใช้ >1 tokens
[ch for ch in thai_chars if ch not in thais]
```

```python
import transformers
import torch

model_id = "deepcogito/cogito-v1-preview-llama-8B"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)

messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Give me a short introduction to LLMs."},
]

outputs = pipeline(
    messages,
    max_new_tokens=512,
)

print(outputs[0]["generated_text"][-1])
```
