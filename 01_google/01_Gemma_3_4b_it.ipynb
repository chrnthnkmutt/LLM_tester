```python
from transformers import pipeline
import torch

pipe = pipeline(
    "image-text-to-text",
    model="google/gemma-3-4b-it",
    device="cuda",
    torch_dtype=torch.bfloat16
)
```

```python
messages = [
    {
        "role": "system",
        "content": [{"type": "text", "text": "You are a helpful assistant."}]
    },
    {
        "role": "user",
        "content": [
            {"type": "image", "url": "https://d2e5ushqwiltxm.cloudfront.net/wp-content/uploads/sites/62/2024/08/29074619/Wat-Phra-Kaew.jpg"},
            {"type": "text", "text": "ที่นี่ที่ไหน?"}
        ]
    }
]

output = pipe(text=messages, max_new_tokens=200)
print(output[0]["generated_text"][-1]["content"])
```

```python
import json
```

```python
data = json.load(open('tokenizer.json'))
```

```python
vocab = data['model']['vocab']
type(vocab), len(vocab)
```

```python
list(vocab.keys())[:20]
```

```python
list(vocab.keys())[259:600]
```

```python
import re
```

```python
pat = re.compile('[ก-๙]')
```

```python
thais = [v for v in vocab.keys() if pat.search(v)]
len(thais)
```

```python
','.join(sorted(thais))
```

```python
# ตัวอักษรไทยที่ควรมี
thai_chars = []
pat2 = re.compile('[ก-ฮะ-ูเ-์๐-๙]')
for i in range(ord('ก'), ord('๙')+3):
  if pat2.search(chr(i)):
    thai_chars.append(chr(i))
len(thai_chars)
```

```python
# ตัวที่ขาดไป  ต้องใช้ >1 tokens
[ch for ch in thai_chars if ch not in thais]
```
